```python
# get the max Tenure 

print(new_df['tenure'].max())
```

```python
# making the labels

labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
print(labels)
```

Convert the target variable Churn into numeric variable


```python
new_df['Churn']=np.where(new_df.Churn=='Yes',1,0)
```

```python
new_df.head()
```

Convert all the category variable into dummies variable

```python
new_df_dummies=pd.get_dummies(new_df,dtype=int)
new_df_dummies.head()
```

```python
sns.lmplot(data=new_df_dummies,x='MonthlyCharges',y='TotalCharges',fit_reg=False)
```

```python
# churn by Monthly charges and totalCharges

Mth=sns.kdeplot(new_df_dummies.MonthlyCharges[(new_df_dummies['Churn']==0)],
               color='Red',shade=True)
Mth=sns.kdeplot(new_df_dummies.MonthlyCharges[(new_df_dummies['Churn']==1)],
               color='Blue',shade=True)
Mth.legend(['NO churn','Churn'],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly Charges by Month')
```

churn is high when monthly charges are high

```python

Tot=sns.kdeplot(new_df_dummies.TotalCharges[(new_df_dummies['Churn']==0)],
               color='Red',shade=True)
Tot=sns.kdeplot(new_df_dummies.MonthlyCharges[(new_df_dummies['Churn']==1)],
               color='Blue',shade=True)
Tot.legend(['NO churn','Churn'],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Monthly Charges')
Tot.set_title('Monthly Charges by Month')
```

Building the correlation of all predictors with churn

```python
plt.figure(figsize=(20,8))
new_df_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```

High churn seen in case of Month to month contract , no online security, first year of subscription and fiber optics internet.

low churn is seen in case of Long term contract , subscription without internet service  and age for 5+yrs
